```
import pandas as pd
df=pd.read_csv('Social_Network_Ads.csv') 
de=df.iloc[:,2:4]   
```

```
de
```

```
import plotly.express as px
fig=px.scatter(de,x=de.iloc[:,0],y=de.iloc[:,1],labels={'x':"Age",'y':'Salary'})
fig.show() 
```

```
from sklearn.cluster import KMeans

km=KMeans(n_clusters=5) 

km.fit(de) 

labels=km.predict(de)
centroids=km.cluster_centers_
```

```
from plotly.validators.scatter.marker import SymbolValidator

symbols = SymbolValidator().values
symbols 
```

```
import plotly.graph_objects as go

fg=px.scatter(de,x=de.iloc[:,0],y=de.iloc[:,1],color=labels,labels={'y':"Salary",'x':'Age'})
cent=pd.DataFrame(centroids,columns=['A','B']) 
fg.add_trace(go.Scatter(x=cent.iloc[:,0],y=cent.iloc[:,1],mode='markers',marker_symbol=symbols[17],marker=dict(color="black",size=12),showlegend=False))  

fg.show() 
```
